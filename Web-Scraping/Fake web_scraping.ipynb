```python
url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)
```

<div style="
background:linear-gradient(135deg,#140022,#050505);
padding:20px;
border-radius:20px;
border-left:8px solid #a855f7;
">
<h2 style="

">
💼 Parsing HTML Content
</h2>
</div>

```python
soup = BeautifulSoup(
    response.text,
    "html.parser"
)
```

<div style="
background:linear-gradient(135deg,#140022,#050505);
padding:20px;
border-radius:20px;
border-left:8px solid #a855f7;
">
<h2 style="
color:#c084fc;
font-size:28px;
">
💼 Extracting Job Details
</h2>
</div>

```python
jobs = soup.find_all(
    "div",
    class_="card-content"
)


print("Total Jobs Found:", len(jobs))
```

```python
job_data = []


for job in jobs:

    title = job.find(
        "h2",
        class_="title"
    ).text.strip()


    company = job.find(
        "h3",
        class_="company"
    ).text.strip()


    location = job.find(
        "p",
        class_="location"
    ).text.strip()


    link = job.find(
        "a"
    )["href"]


    job_data.append({

        "Title": title,

        "Company": company,

        "Location": location,

        "Apply Link": link

    })
```
