```python
import requests
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
response = requests.get(url)
```

<h2 style="color:#4F46E5; margin-top:0;">

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content , "html.parser")
soup
```

```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all("div", class_="thumbnail")
soup1
```

<div style="
background:#F8F9FF;
border-left:8px solid #4F46E5;
padding:20px;
border-radius:10px;
">

<h2 style="color:#4F46E5; margin-top:0;">
💻 Fetching the Data
</h2>


</div>

```python
data = []
```

```python
for product in soup1:

    title = product.find("a", class_="title").get("title")
    price = product.find("h4", class_="price").text.strip()
    description = product.find("p", class_="description").text.strip()

    data.append([title, price, description])
```

```python
len(soup1)
```

```python
df = pd.DataFrame(data, columns=["Title", "Price", "Description"])

df
```

```python
df.to_csv("laptop_products.csv", index=False)
```
