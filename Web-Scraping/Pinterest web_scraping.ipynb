<h2 style="color:#d63384;">
🐥 Importing Libraries
</h2>
</div>

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup
import pandas as pd
import time
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Selenium Browser Setup
</h2>
</div>

```python
options = Options()

options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Search Keyword Input
</h2>
</div>

```python
keyword = input("Enter Pinterest search keyword: ")

print("Searching for:", keyword)
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Open Pinterest Search Page
</h2>
</div>

```python
search_url = f"https://www.pinterest.com/search/pins/?q={keyword}"

driver.get(search_url)

time.sleep(5)

print("Current URL:", driver.current_url)
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Extract Page Source
</h2>
</div>

```python
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Scroll Page for Dynamic Loading
</h2>
</div>

```python
for i in range(5):
    driver.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);"
    )
    time.sleep(2)
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);
padding:20px;
border-radius:20px;
border-left:8px solid #ffb6c1;
">
<h2 style="color:#d63384;">
🐥 Extract Image URLs
</h2>
</div>

```python
image_urls = []

images = driver.find_elements(By.TAG_NAME, "img")

for i in range(len(images)):
    try:
        images = driver.find_elements(By.TAG_NAME, "img")  # refresh elements
        
        src = images[i].get_attribute("src")

        if src and src.startswith("http"):
            image_urls.append(src)

    except Exception:
        pass


print("Total Images Found:", len(image_urls))

for url in image_urls[:5]:
    print(url)
```

<div style="
background:linear-gradient(135deg,#fff5f8,#ffeef2);

">
<h2 style="color:#d63384;">
🐥 Creating DataFrame
</h2>
</div>

```python
pinterest_data = pd.DataFrame({
    "Image No": range(1, len(image_urls) + 1),
    "Image URL": image_urls
})

pinterest_data
```

<h2 style="color:#d63384;">
🐥 Display Scraped Images
</h2>

```python
from IPython.display import Image, display
import requests
from io import BytesIO


for url in image_urls[:5]:
    try:
        response = requests.get(url)
        img = Image(data=response.content)
        display(img)

    except Exception:
        print("Image could not be loaded")
```
