```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content , "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all("h3")
```

 👽 Fetching the data
</h3>

</div>

```python
data = []

for item in soup1:
    title = item.get_text(strip=True)

    data.append({
        'Title': title
    })

data
```
