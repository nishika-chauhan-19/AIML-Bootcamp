<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Importing The Library</b>
</div>

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import pandas as pd
```

<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Launching Chrome Browser</b>
</div>

```python
options = Options()

options.add_argument("--headless=new")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)


keyword = "sunscream"

driver.get(
    f"https://www.amazon.in/s?k={keyword}"
)
```

<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Waiting for the Webpage to Load</b>
</div>

```python
import time

time.sleep(5)
```

<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Extracting Page Source</b>
</div>

```python
html = driver.page_source
```

<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Parsing HTML using BeautifulSoup</b>
</div>

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
products = soup.find_all("div", attrs={"data-component-type": "s-search-result"})
len(products)
```

<div style="background:#FAF4EA;padding:5px 10px;border-left:5px solid #8B5E3C;">
<b>★ Fetch the data</b>
</div>

```python
data = []

for product in products:

    try:
        title = product.find("h2").text.strip()
    except:
        title = "N/A"

    try:
        price = product.find("span", class_="a-price-whole").text.strip()
    except:
        price = "N/A"

    try:
        rating = product.find("span", class_="a-icon-alt").text.strip()
    except:
        rating = "N/A"

    data.append([title, price, rating])
```

```python
df = pd.DataFrame(data, columns=["Title", "Price", "Rating"])
df
```

```python
df.to_csv("amazon_mobile_data.csv", index=False)
```
