```python
import requests
url = "https://www.worldometers.info/world-population/"
response = requests.get(url)
```

<h2 style="color:#065F46; margin-top:0;">
🌍 Parsing the Content 
</h2>



</div>

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content , "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all("table")
soup1
```

<div style="
background:#F0FDF4;
border-left:8px solid #065F46;
padding:20px;
border-radius:10px;
">

<h2 style="color:#065F46; margin-top:0;">
🌍 Fetching the data 
</h2>



</div>

```python
data = []


table = soup.find("table")


rows = table.find_all("tr")


for i, row in enumerate(rows[1:], start=1):

    columns = row.find_all("td")

    if columns:
        country = columns[1].text.strip()
        population = columns[2].text.strip()

        data.append([i, country, population])


df = pd.DataFrame(data, columns=["S.No", "Country", "Population"])

df
```

```python
df.to_csv("world_population.csv", index=False)
```
