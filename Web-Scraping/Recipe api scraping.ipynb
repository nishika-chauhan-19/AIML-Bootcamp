<h2 style="color:#4E342E;margin-top:0;">
🍽️ Importing the library
</h2>

```python
import requests
import pandas as pd
```

<h2 style="color:#4E342E;margin-top:0;">
🍽️ Define API Endpoint
</h2>

```python
url = "https://dummyjson.com/recipes"
```

<div style="
background:#FFFFFF;

border-radius:12px;
border:1px solid #D7CCC8;

">
<h2 style="color:#4E342E;margin-top:0;">
🍽️ Send Request to the API
</h2>

```python
response = requests.get(url)
```

background:#FFFFFF;

```python
response.status_code
```

<div style="
background:#FFFFFF;

<h2 style="color:#4E342E;margin-top:0;">
🍽️ Convert Response to JSON
</h2>

```python
data = response.json()

print(type(data))
print(data.keys())

data
```

<div style="
background:#FFFFFF;
padding:20px;
border-left:8px solid #6D4C41;
border-radius:12px;
border:1px solid #D7CCC8;
font-family:'Segoe UI',sans-serif;
margin-top:8px;
">
<h2 style="color:#4E342E;margin-top:0;">
🍽️  Data extraction
</h2>

```python
recipe_data = []

for recipe in data["recipes"]:
    recipe_data.append({
        "ID": recipe["id"],
        "Recipe Name": recipe["name"],
        "Cuisine": recipe["cuisine"],
        "Meal Type": ", ".join(recipe["mealType"]),
        "Prep Time (min)": recipe["prepTimeMinutes"],
        "Cook Time (min)": recipe["cookTimeMinutes"],
        "Servings": recipe["servings"],
        "Difficulty": recipe["difficulty"],
        "Calories": recipe["caloriesPerServing"],
        "Rating": recipe["rating"],
        "Reviews": recipe["reviewCount"],
        "Ingredients": ", ".join(recipe["ingredients"]),
        "Tags": ", ".join(recipe["tags"]),
    })
recipe_data    
```

<h2 style="color:#4E342E;margin-top:0;">
🍽️  Creating DataFrame 
</h2>

```python
df = pd.DataFrame(recipe_data)

df.head()
```

</h2>

```python
df.to_csv("recipes_data.csv", index=False)
```
