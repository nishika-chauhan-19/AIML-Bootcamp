<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Import Required Libraries
</h2>
</div>

```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Configure Chrome options
</h2>
</div>

```python
options = Options()
options.add_argument("--start-maximized")
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Launch Browser & Access GitHub Trending Page
</h2>
</div>

```python
driver = webdriver.Chrome(options=options)
url = "https://github.com/trending"
driver.get(url)
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Wait for page loading
</h2>
</div>

```python
time.sleep(5)
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Page Source Extract & HTML Parsing
</h2>
</div>

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Display page title
</h2>
</div>

```python
print("Page Title:", soup.title.text if soup.title else "No title found")
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 Repository Data Extraction
</h2>
</div>

```python
repo_cards = soup.find_all("article", class_="Box-row")
repo_names = []
descriptions = []
languages = []
stars = []
repo_links = []
for repo in repo_cards:
    

    name = repo.h2.get_text(strip=True).replace("\n", "")
    repo_names.append(name)
    
    
    desc = repo.find("p")
    descriptions.append(desc.get_text(strip=True) if desc else "No description available")
    
    
    lang = repo.find("span", itemprop="programmingLanguage")
    languages.append(lang.get_text(strip=True) if lang else "Not specified")
    
    
    star = repo.find("a", href=lambda x: x and "/stargazers" in x)
    stars.append(star.get_text(strip=True) if star else "0")
    
    
    link = repo.h2.find("a")
    repo_links.append("https://github.com" + link["href"].strip() if link else "")
print("Total repositories found:", len(repo_names))
```

<div style="background-color:#0d1117; padding:20px; border-radius:15px; border:3px solid #2ea043;">

<h2 style="color:#2ea043; text-shadow:0 0 10px #2ea043;">
🐙 DataFrame Creation
</h2>
</div>

```python
data = {
    "Repository": repo_names,
    "Description": descriptions,
    "Language": languages,
    "Stars": stars,
    "URL": repo_links
}
df = pd.DataFrame(data)
df
```

```python
df.to_csv("github_trending_repositories.csv", index=False)
```
