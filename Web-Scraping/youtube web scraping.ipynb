★ Searching for Videos

```python
search = driver.find_element(By.NAME, "search_query")
search.send_keys("Python Tutorial")
search.send_keys(Keys.ENTER)

time.sleep(5)

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

<b style="color:#8B1E1E;font-size:15px;font-family:Georgia,serif;">
★ Finding Video Containers
</b>

</div>

```python
videos = soup.find_all("ytd-video-renderer")
len(videos)
```

<div style="

<b style="color:#8B1E1E;font-size:15px;font-family:Georgia,serif;">
★ Extracting Video Details
</b>

</div>

```python
data = []

for video in videos:

    try:
        title = video.find("a", id="video-title").text.strip()
    except:
        title = "N/A"

    try:
        channel = video.find("a", class_="yt-simple-endpoint style-scope yt-formatted-string").text.strip()
    except:
        channel = "N/A"

    try:
        meta = video.find("div", id="metadata-line")
        spans = meta.find_all("span")

        views = spans[0].text.strip()
        upload = spans[1].text.strip()

    except:
        views = "N/A"
        upload = "N/A"

    data.append([title, channel, views, upload])
```
