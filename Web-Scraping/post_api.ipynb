🚀 API Endpoint
</h2>
</div>

```python
base_url = "https://jsonplaceholder.typicode.com"
endpoint = "/posts"
api_url = base_url + endpoint
print("API URL:", api_url)
```

<div style="
background: linear-gradient(135deg,#111827,#1E293B);
padding:35px;
border-radius:30px;
border:1px solid #64748B;
box-shadow:0 10px 35px rgba(0,0,0,0.4);
">
<h2 style="
color:#E0E7FF;
font-size:20px;
font-weight:700;
">
🚀 Sending GET Request
</h2>
</div>

```python
response = requests.get(api_url)
print("Status Code:", response.status_code)
```

<div style="
background: linear-gradient(135deg,#111827,#1E293B);
padding:35px;
border-radius:30px;
border:1px solid #64748B;
box-shadow:0 10px 35px rgba(0,0,0,0.4);
">
<h2 style="
color:#E0E7FF;
font-size:20px;
font-weight:700;
">
🚀 Convert JSON Response
</h2>
</div>

```python
data = response.json()
data[0]
```

<div style="
background: linear-gradient(135deg,#111827,#1E293B);
padding:35px;
border-radius:30px;
border:1px solid #64748B;
box-shadow:0 10px 35px rgba(0,0,0,0.4);
">
<h2 style="
color:#E0E7FF;
font-size:20px;
font-weight:700;
">
🚀 Data extraction
</h2>
</div>

```python
posts_list = []

for post in data:
    posts_list.append({
        "User ID": post["userId"],
        "Post ID": post["id"],
        "Title": post["title"],
        "Body": post["body"]
    })

print("Total Posts Extracted:", len(posts_list))
```

">
🚀Dataframe
</h2>
</div>

```python
df = pd.DataFrame(posts_list)

df
```
