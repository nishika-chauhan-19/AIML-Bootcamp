<div style="background:#fbf8fc;padding:5px 10px;border-left:3px solid #a88bbf;">
<b>✦ Extracting of Content</b>
</div>

```python
import requests
url = "https://en.wikipedia.org/wiki/Comparison_of_programming_languages"
response = requests.get(url)
```

<div style="background:#fbf8fc;padding:5px 10px;border-left:3px solid #a88bbf;">
<b>✦ Checking the Status Code</b>
</div>

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
response = requests.get(url, headers=headers)
```

```python
print(response.status_code)
```

<div style="background:#fbf8fc;padding:5px 10px;border-left:3px solid #a88bbf;">
<b>✦ Parsing the Content</b>
</div>

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content , "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all("table")
soup1
```

<div style="background:#fbf8fc;padding:5px 10px;border-left:3px solid #a88bbf;">
<b>✦ Fetching the Data</b>
</div>

```python
soup1 = []
```

```python
for item in soup.find_all("tr"):
    data = item.find_all("td")

    if len(data) > 0:
        row = [i.get_text(strip=True) for i in data]

        soup1.append(row)
```

```python
soup1
```

```python
df = pd.DataFrame(soup1)
df
```

df.to_CSV("wikipedia", index=False)
