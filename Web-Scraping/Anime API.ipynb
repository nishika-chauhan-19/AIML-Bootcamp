🎌 Data extraction
</div>

```python
anime_list = []

for anime in data["data"]:
    anime_list.append({
        "Title": anime["title"],
        "Score": anime["score"],
        "Rank": anime["rank"],
        "Episodes": anime["episodes"],
        "Popularity": anime["popularity"],
        "Status": anime["status"],
        "Year": anime["year"]
    })

anime_list
```
