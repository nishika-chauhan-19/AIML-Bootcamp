<h2 style="color:#8e44ad;">

```python
api_url = "https://api.spaceflightnewsapi.net/v4/articles/?limit=10"

print("API Endpoint:")
print(api_url)
```

</h2>


</div>

```python
response = requests.get(api_url)
response
```

<h2 style="color:#8e44ad;">

</h2>

```python
news_data = response.json()
news_data
```

<div style="

<h2 style="color:#8e44ad;">
🌌 Data extraction   
</h2>


</div>

```python
articles = news_data["results"]

extracted_data = []

for article in articles:
    
    data = {
        "Title": article["title"],
        "Summary": article["summary"],
        "Published Date": article["published_at"],
        "URL": article["url"]
    }
    
    extracted_data.append(data)


extracted_data[:2]
```

<h2 style="color:#8e44ad;">
🌌 Creating DataFrame   
</h2>


</div>

```python
space_news_df = pd.DataFrame(extracted_data)
space_news_df
```
