◇ Creating the Synthetic Dataset
</h4>

```python
import numpy as np 
import pandas as pd 
np.random.seed(54)
n_student = 10000

data = {
  
    "name" : np.random.choice(["DEV","HARSHU","NISHU","HARSHIKA"], n_student),
    "age" : np.random.randint(18, 25, n_student),
    "city" : np.random.choice(["Ambala","radaur","yamunanagar","ladwa","delhi"])
}
df = pd.DataFrame(data)
df
```
