```python
import numpy as np
np.random.seed(42)
n_students = 100

data = {
    "student_name": np.random.choice(
        ["Aarav","Vivaan","Ananya","Diya","Rahul","Priya","Riya","Harsh","Neha","Yash"],
        n_students
    ),
    "gender": np.random.choice(
        ["Male","Female"],
        n_students
    ),
    "age": np.random.randint(
        18,26,n_students
    ),
    "department": np.random.choice(
        ["Computer Science","Data Science","Artificial Intelligence","Information Technology","Electronics"],
        n_students
    ),
    "semester": np.random.randint(
        1,9,n_students
    ),
    "study_hours": np.round(
        np.random.uniform(1,8,n_students),1
    ),
    "attendance": np.round(
        np.random.uniform(60,100,n_students),1
    ),
    "assignment_score": np.random.randint(
        50,101,n_students
    ),
    "internal_marks": np.random.randint(
        10,31,n_students
    ),
    "exam_score": np.random.randint(
        35,101,n_students
    )
}
df = pd.DataFrame(data)
def grade(score):
    if score >= 90:
        return "A+"
    elif score >= 80:
        return "A"
    elif score >= 70:
        return "B+"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    else:
        return "F"
df["grade"] = df["exam_score"].apply(grade)
df["result"] = df["exam_score"].apply(
    lambda x: "Pass" if x >= 40 else "Fail"
)
df
```
