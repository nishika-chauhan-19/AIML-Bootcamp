## Question 1
What is the difference between == and is in Python?

Answer:
The == operator is used to compare the values of two objects, whereas the is operator is used to check whether two variables refer to the same object in memory.

## Question 8
Why should you never use str += inside a loop? What should you use instead?

Answer:
Using str += inside a loop is inefficient because strings are immutable, so a new string is created every time the loop runs.

````python
result = ""
for i in range(5):
    result += str(i)

print(result)
````

````python
#Better methood
result = "".join(str(i) for i in range(5))
print(result)
````

````python
# EC-1: EMPTY INPUT CASES
# The function below works for normal input — make it handle all edge cases
def average(numbers):
 """Return the average of a list of numbers."""
 return sum(numbers) / len(numbers)
# Test cases to handle:
# average([]) → should return None or raise a clear ValueError
# average([5]) → should return 5.0
# average(None) → should raise TypeError with clear message
# average(["a", "b"]) → should raise TypeError
# average([1, 2, None, 4]) → handle None in list

Answer:
improve version :
def average(numbers):
    """Return the average of a list of numbers."""

    if numbers is None:
        raise TypeError("Input cannot be None")

    if len(numbers) == 0:
        raise ValueError("Cannot calculate average of empty list")

    for i in numbers:
        if i is None:
            raise TypeError("List contains None value")
        if not isinstance(i, (int, float)):
            raise TypeError("All elements must be numbers")

    return sum(numbers) / len(numbers)


print(average([]))
print(average([5]))
print(average(None))
print(average(["a", "b"]))
print(average([1, 2, None, 4]))
````

````python
# EC-2: BOUNDARY CONDITIONS
# Check if a password meets requirements:
# Length: 8-20 chars, has uppercase, lowercase, digit, special char
def is_valid_password(pwd):
 # Test these boundaries:
 # "" → False (empty)
 # "abc" → False (too short)
 # "a" * 21 → False (too long)
 # "Abcdefg1!" → True
 # "ABCDEFG1!" → False (no lowercase)
 # "abcdefg1!" → False (no uppercase)
 # "Abcdefgh!" → False (no digit)
 # "Abcdefg12" → False (no special char)
 pass

Answer:
improve version:
def is_valid_password(pwd):
    if len(pwd) < 8 or len(pwd) > 20:
        return False

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for ch in pwd:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        else:
            has_special = True

    return has_upper and has_lower and has_digit and has_special
````

````python
# B1: Write a function that takes a temperature in Celsius and returns it in
# Fahrenheit AND Kelvin. Return as a tuple: (fahrenheit, kelvin)
# Formula: F = (C × 9/5) + 32 | K = C + 273.15

def convert_temperature(celsius):
    fahrenheit = (celsius * 9/5) + 32
    kelvin = celsius + 273.15
    return (fahrenheit, kelvin)
````

````python
# B2: Write a function that checks if a string is a palindrome
# Ignore case and spaces
# "A man a plan a canal Panama" → True
# "hello" → False

def is_palindrome(text):
    text = text.replace(" ", "").lower()
    return text == text[::-1]
````
