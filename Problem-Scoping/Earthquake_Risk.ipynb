```python
import numpy as np
import pandas as pd

np.random.seed(54)
n_regions = 1000

data = {
    "Region_ID": [f"REG{i:05d}" for i in range(1, n_regions+1)],
    
    "Seismic_Activity": np.random.randint(1, 10, n_regions),
    "Fault_Line_Distance_km": np.random.randint(1, 200, n_regions),
    "Ground_Movement_mm": np.random.randint(0, 100, n_regions),
    "Soil_Stability_Index": np.random.randint(1, 10, n_regions),
    "Population_Density": np.random.randint(100, 10000, n_regions),
    "Building_Vulnerability": np.random.randint(1, 10, n_regions),
    "Historical_Earthquakes": np.random.randint(0, 20, n_regions),
    "Emergency_Preparedness": np.random.randint(1, 10, n_regions),
    "Sensor_Alerts": np.random.randint(0, 15, n_regions)
}

df = pd.DataFrame(data)
#print(data)
#print(df)
len(df)
```

👁️ Calculating the risk score
</h2>

</div>

```python
#Risk score based on Domain Knowledge
#Calculate weighted Risk Score

risk_score = (
    0.25 * (df["Seismic_Activity"] > 6).astype(int) +
    0.20 * (df["Fault_Line_Distance_km"] < 50).astype(int) +
    0.15 * (df["Ground_Movement_mm"] > 60).astype(int) +
    0.10 * (df["Soil_Stability_Index"] < 4).astype(int) +
    0.10 * (df["Population_Density"] > 5000).astype(int) +
    0.10 * (df["Building_Vulnerability"] > 6).astype(int) +
    0.05 * (df["Historical_Earthquakes"] > 10).astype(int) +
    0.05 * (df["Sensor_Alerts"] > 8).astype(int)
)

risk_score += np.random.normal(0, 0.05, len(df))
```
