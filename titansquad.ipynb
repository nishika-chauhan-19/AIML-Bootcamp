```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content , "html.parser")
soup
```

```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all('td', class_="title")
soup1
```

<div style="background:#f5f7ff;padding:5px 10px;border-left:3px solid #4a5cff;">
<b>🔹Fetching the Data</b>
</div>

```python
soup1 = []
```

```python
for item in soup.find_all("tr", class_="athing"):
    News_name = item.find("span", class_="titleline").get_text(strip=True)
    Link = item.find("a")["href"]

    soup1.append({
        "News_name": News_name,
        "Link": Link
    })
```

```python
soup1
```

```python
df = pd.DataFrame(soup1)
df
```

df.to_CSV("news_data", index=False)
